```python
# Read JSON file into a pandas DataFrame

import pandas as pd

file_path2 = r'/gdrive/My Drive/CIS509 Project/yelp_academic_dataset_review.json'
chunk_size = 1000000  # Set the desired chunk size

# Create an iterator to read the file in chunks
json_reader = pd.read_json(file_path2, lines=True, chunksize=chunk_size)

# List to store individual DataFrames
dfs = []

# Iterate over chunks and append each chunk to the list
for i, chunk in enumerate(json_reader):
    print(f"Processing Chunk {i + 1}, Shape: {chunk.shape}")
    dfs.append(chunk)

# Concatenate all DataFrames in the list into one DataFrame
full_df = pd.concat(dfs, ignore_index=True)

# Display the shape of the concatenated DataFrame
print("Final DataFrame Shape:", full_df.shape)
```

```python
# Null values

print(full_df.isnull().sum())
```

```python
# Filter the review dataset to 2018 and 2022 reviews respectively (So we can join the columns)

# Extract and display rows for the years 2018 and 2021
df_filtered = full_df[(full_df['date'].dt.year == 2018) | (full_df['date'].dt.year == 2021)]

# Display the resulting DataFrame

df_filtered.tail()
```

```python
df_filtered.shape
```

```python
# Detect common cols

# Inspect DataFrame columns
columns_df1 = set(df.columns)
columns_df2 = set(df_filtered.columns)

# Find common columns
common_columns = columns_df1.intersection(columns_df2)

# Display common columns
print("Common Columns:", common_columns)
```

```python
# Join the 2 json files

# Assuming df_review and df_other are your DataFrames
# 'business_id' is the common column

# Merge the DataFrames based on the common column 'business_id'
merged_df = pd.merge(df, df_filtered, on = "business_id", how="inner")

# Display the resulting joined DataFrame
merged_df.tail(3)
```

## **Retaining ONLY restaurants**

```python
# Filtering ONLY restaurants

import pandas as pd

# Assuming df is your DataFrame and 'category_column' is the name of the category column
# Replace 'terms' with the specific terms you want to search for
terms = ['Restaurants', 'Food', 'Bubble Tea', 'Coffee & Tea', 'Bakeries','Salad',
         'Coffee & Tea', 'Breakfast & Brunch','Sandwiches', 'Sports Bars','Nightlife','Burgers','Cafes',
         'Pizza', 'Italian','Coffee Roasteries','Thai','Japanese', 'American Food', 'Mexican','Vegan','Fast Food', 'Vegan', 'Vegetarian',
         'Delis', 'Vietnamese', 'Pie']

# Count the occurrences of each term in each row
occurrences = merged_df2['categories'].str.count('|'.join(terms))

# Create a boolean mask to filter rows containing at least three of the specified terms
mask = occurrences >= 2

# Filter the DataFrame using the mask
filtered_df = merged_df2[mask]

# Print the filtered DataFrame
print(filtered_df.shape)
```
